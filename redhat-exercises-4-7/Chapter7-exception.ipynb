```python
# EXERCISE 1
def main():
    numeros = [4-13]
    
    while True:
        entrada = input("Enter a number or 'end' to quit: ")
        
        if entrada == "end":
            break
            
        try:
            indice = int(entrada)
            print(numeros[indice])
        except ValueError:
            print("Illegal number. Please enter a valid integer.")
        except IndexError:
            print("Illegal subscript. Index out of range.")

if __name__ == "__main__":
    main()
```

```python
# EXERCISE 2
def main():
    numeros = [4-13]
    
    while True:
        entrada = input("Enter a number or 'end' to quit: ")
        
        if entrada == "end":
            break
            
        try:
            indice = int(entrada)
            
            if indice < 0:
                raise IndexError("Negative numbers are not allowed")
                
            print(numeros[indice])
        except ValueError:
            print("Illegal number. Please enter a valid integer.")
        except IndexError:
            print("Illegal subscript. Index out of range or negative.")

if __name__ == "__main__":
    main()
```

```python
# EXERCISE 3
def main():
    total = 0
    
    while True:
        try:
            entrada = input("Enter an integer: ")
            total += int(entrada)
        except ValueError:
            print("ValueError: That is not a valid integer.")
        except KeyboardInterrupt:
            print("\nKeyboardInterrupt: Program terminated by user.")
            break
        except EOFError:
            print("\nEOFError: End of stream reached.")
            break
            
    print("Sum of all integers entered:", total)

if __name__ == "__main__":
    main()
```
