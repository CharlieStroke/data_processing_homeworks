```python
# EXERCISE 1
def validar_entero_positivo():
    entrada = input("Por favor, ingresa un número entero positivo: ")
    
    # (Nota: el uso de isdigit() es una práctica estándar en Python para validar strings)
    if entrada.isdigit() and int(entrada) > 0:
        return int(entrada)
    else:
        return 0

while True:
    resultado = validar_entero_positivo()
    
    if resultado == 0:
        print("Mensaje de la aplicación: Entrada inválida. Inténtalo de nuevo.\n")
    else:
        print(f"¡Correcto! El número válido ingresado es: {resultado}")
        break
```

```python
# EXERCISE 2
def longitud_maxima(coleccion):
    max_len = 0
    for cadena in coleccion:
        if len(cadena) > max_len:
            max_len = len(cadena)
    return max_len

palabras = ["Hola", "Desarrollador", "Python", "Funciones", "Solución"]

ancho_maximo = longitud_maxima(palabras)

print("Palabras justificadas a la derecha:")
for palabra in palabras:
    formato_string = "{:>" + str(ancho_maximo) + "}"
    print(formato_string.format(palabra))
```

```python
# EXERCISE 3
def mi_suma(*args):
    total = 0
    for numero in args:
        total += numero
    return total

resultado_1 = mi_suma(1, 2, 3, 4, 5)
print("La suma de 1, 2, 3, 4, 5 es:", resultado_1)

resultado_2 = mi_suma(5, 2, 7)
print("La suma de 5, 2, 7 es:", resultado_2)
```

```python
# EXERCISE 4
def suma_y_promedio(*args):
    if len(args) == 0:
        return (0, 0)
        
    total = sum(args)
    cantidad = len(args)
    promedio = total / cantidad
    
    return (total, promedio)

resultado = suma_y_promedio(10, 20, 30, 40)
print("El resultado devuelto es:", resultado)
print("El tipo de dato devuelto es:", type(resultado))

suma_total, valor_promedio = suma_y_promedio(5, 2, 7)
print(f"Para los valores 5, 2, 7 -> Suma: {suma_total}, Promedio: {valor_promedio}")
```

```python
# EXERCISE 5
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: No se puede dividir por cero"
    return a / b

def quit_app():
    print("Terminando la aplicación...")
    exit()

def illegal():
    print("Selección inválida. Por favor, intenta de nuevo.\n")

menu = {
    "1": add,
    "2": subtract,
    "3": multiply,
    "4": divide
}

while True:
    print("\nCalculator options:")
    print("      1. Add")
    print("      2. Subtract")
    print("      3. Multiply")
    print("      4. Divide")
    print("      5. Quit")
    
    choice = input("Enter a number from the above menu: ")
    
    if choice == "5":
        quit_app()
    elif choice in menu:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        
        resultado = menu[choice](num1, num2)
        print("Result:", resultado)
    else:
        illegal()
```

```python
# EXERCISE 6
def get_positive_elements(numbers):
    return list(filter(lambda x: x > 0, numbers))

test_list = [-10, 15, -3, 0, 7, 22, -1]
result = get_positive_elements(test_list)
print(result)
```

```python
# EXERCISE 7
def positive(*args, num):
    count = 0
    for value in args:
        if value > num:
            count += 1
    return count

res = positive(5, -10, 10, -20, 30, num=0)
print(res)
```

```python
# EXERCISE 11
def get_common_elements(list1, list2):
    common = []
    for item in list1:
        if item in list2 and item not in common:
            common.append(item)
    return common

list_a = [1, 3 , 5]
list_b = [2, 3, 5, 6]
result = get_common_elements(list_a, list_b)
print(result)
```

```python
# EXERCISE 12
def add_to_values(num, data_dict):
    for key in data_dict:
        data_dict[key] += num
    return data_dict

sample_dict = {'a': 10, 'b': 20, 'c': 30}
result = add_to_values(5, sample_dict)
print(result)
```

```python
# EXERCISE 13
def find_occurrence(search_list, target, occurrence):
    count = 0
    for index, item in enumerate(search_list):
        if item == target:
            count += 1
            if count == occurrence:
                return index
    raise ValueError("El valor no alcanzó el número de ocurrencias buscado o no existe en la lista")

data_list = [2, 3, 4, 5]
result = find_occurrence(data_list, 20, 3)
print(result)
```
