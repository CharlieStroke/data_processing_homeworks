```python
logs_crudos = [
    ("192.168.1.1", "ERROR", "Timeout"),
    ("10.0.0.5", "INFO", "Login exitoso"),
    ("192.168.1.1", "ERROR", "Timeout"),
    ("172.16.0.10", "WARNING", "Memoria alta"),
    ("10.0.0.5", "INFO", "Login exitoso"),
    ("192.168.1.1", "ERROR", "Conexión rechazada")
]

logs_unicos = set(logs_crudos)

print(f"Eventos únicos detectados: {len(logs_unicos)}")

servidor_master = ("192.168.1.1", 9200)

# si se intenta hacer esto: servidor_master[1] = 9300, el compilador dara error dado a que las tuplas son inmutables

ips_detectadas = []
for log in logs_unicos:
    ips_detectadas.append(log[0])

print(f"Lista de IPs extraídas: {ips_detectadas}")

reporte = {
    "ip": servidor_master[0],
    "eventos_unicos": len(logs_unicos),
    "critico": len(logs_unicos) > 2
}

print("REPORTE DE SEGURIDAD:")
print(reporte)
```
